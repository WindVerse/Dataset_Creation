# Derive displacements

```python
dataset_ver = 4

num_of_iterations = 100
frames_per_iteration = 300
```

```python
import os
import shutil
import numpy as np
```

```python
dataset_path = "./../../datasets/5/flags"
output_path = "./../../datasets/5/displacements"

if os.path.exists(output_path):
    shutil.rmtree(output_path)

os.makedirs(output_path, exist_ok=False)
```

```python

for i in range (1, num_of_iterations+1):
  for j in range (0, frames_per_iteration):
    first = np.load(dataset_path + f'/flag_{i:03d}_{j:03d}.npy')
    second = np.load(dataset_path + f'/flag_{i:03d}_{j+1:03d}.npy')
    displacement = second - first
    np.save(output_path + f'/displacement_{i:03d}_{j:03d}.npy', displacement)
print(f"displacements saved to {output_path}")
```

```python
first = np.load(dataset_path+'/flag_001_000.npy')
second = np.load(dataset_path+'/flag_001_001.npy')
displacement = np.load(output_path+'/displacement_001_000.npy')

print("First frame (10 vertices):\n", first[:10])
print("\nSecond frame (10 vertices):\n", second[:10])
print("\nDisplacement (10 vertices):\n", displacement[:200])
```
