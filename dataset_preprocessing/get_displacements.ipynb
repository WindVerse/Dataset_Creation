# Derive displacements

```python
dataset_ver = 3
```

```python
from google.colab import drive
import zipfile
import os
import shutil
import numpy as np
```

```python
drive.mount('/content/drive')
```

```python
zip_path = "/content/drive/MyDrive/FYP/datasets/"+str(dataset_ver)+"/BlenderOutputs_Sim.zip"

extract_dir = "/content/dataset"
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("✅ Extraction complete. Files are in:", extract_dir)
```

```python
num_of_iterations = 100
frames_per_iteration = 300
```

```python
dataset_path = "/content/dataset/BlenderOutputs_Sim"
output_path = "/content/dataset/target"

if os.path.exists(output_path):
    shutil.rmtree(output_path)

os.makedirs(output_path, exist_ok=False)

for i in range (1, num_of_iterations + 1):
  for j in range (1, frames_per_iteration):
    first = np.load(dataset_path + f'/flag_{i:03d}_{j:03d}.npy')
    second = np.load(dataset_path + f'/flag_{i:03d}_{j+1:03d}.npy')
    displacement = second - first
    np.save(output_path + f'/displacement_{i:03d}_{j:03d}.npy', displacement)
print(f"displacements saved to {output_path}")
```

```python
shutil.make_archive("displacements", 'zip', output_path)
```
